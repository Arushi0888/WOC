```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_df=pd.read_csv('Lineardata_train.csv')
```

```python
data_df
```

```python
X_features=data_df.drop(['target'],axis=1).copy()
Y_target=data_df['target'].copy()
```

```python
X_features["intercept"]=1
```

```python
X_features
```

```python
Y_target
```

```python
X_features_T=X_features.T
```

```python
X_features_T
```

```python
B=np.linalg.inv(X_features_T@X_features)@X_features_T@Y_target
```

```python
B
```

```python
data=pd.read_csv('Lineardata_test.csv')
```

```python
data
```

```python
X=data.drop(['ID'],axis=1).copy()
Y=data['ID'].copy()
```

```python
X["intercept"]=1
```

```python
X
```

```python
Y
```

```python
B.index=X.columns
```

```python
predictions=X@B
```

```python
predictions
```

```python
SSR=((Y-predictions)**2).sum()
```

```python
SST=((Y-Y.mean())**2).sum()
```

```python
SSR
```

```python
SST
```

```python
R2=1-(SSR/SST)
```
